#### 3. Correlation heatmap

#### 3.1. Heatmap with matplotlib

```python
# Firstly I will create a heatmap with matplotlib with all the columns that we kept

plt.matshow(df.corr())
plt.show()
```

This visualization doesn't tell me anything at the moment, I will create a subgroup for more relevancy

```python
# We are going to do a subgroup for our heatmap, including just index, region, average prices and sales volumes
df_sub = df[['Index','RegionName','AveragePrice','SalesVolume','DetachedPrice','SemiDetachedPrice','TerracedPrice','FlatPrice','CashPrice','MortgagePrice','FTBPrice','NewPrice','OldPrice','NewSalesVolume','OldSalesVolume']]
```

```python
plt.matshow(df_sub.corr())
plt.show()
```

Again this visualization doesn't give us much information,let's add some value to it

```python
# In order to make our heatmap more understandable and clear, lets change some features

plt.title('Correlation Matrix', fontsize=14) # To add a title to our heatmap
f = plt.figure(figsize=(10,10)) # To change figure size
plt.matshow(df_sub.corr(), fignum=f.number) # To change type of plot
plt.xticks(range(df_sub.shape[1]), df_sub.columns, fontsize=10, rotation=75) # To define the x axis labels
plt.yticks(range(df_sub.shape[1]), df_sub.columns, fontsize=10) # TO define the y axis labels
cb = plt.colorbar() # To add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # To add font size to the colorbar
```

The color bar only goes to slightly under 0, meaning that no variables have a strong negative correlation

#### 3.2. Heatmap with seaborn

```python
# To check how our dataframe with selected values look like

df_sub.head()
```

```python
# We have to create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_sub.corr(), annot = True, ax = ax)
# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

The heatmap with seaborn is easier to interpret. The areas with black shade represent variables that are not correlated at all or have a negative weak correlation. 
The areas in salmon color show a positive correlation.
The areas with the lightest shade show a strong positive correlation. This means that the variables that cross in that box increase or decrease in the same direction.
All of this make sense as we are talking about house prices, and when house prices increase, they do it for every type of house. 

#### 4. Scatterplot

```python
# Firstly we create a scatterplot for the "Old Price" and "New Price" columns in seaborn

sns.lmplot(x = 'OldPrice', y = 'NewPrice', data = df_sub)
```

As we can see and we predicted before, the correlation between prices is strong and positive, as when market goes up, it does it as a whole 

```python
# Now, I am going to compare two non correlated like "Sales Volume" and "Average Price"

sns.lmplot(x = 'SalesVolume', y = 'AveragePrice', data = df_sub)
```

The trend line for this variables is very flat. We can observe that, as we could expect, a higher average price means less Sales Volume.

#### 6. Categorical plot

```python
# Initially we create a categorical variable that splits the "Average Price" column into categories.
# For this, we need to create a histogram first to determine the categories

sns.histplot(df['AveragePrice'], bins = 20, kde = True)
```

```python
df.loc[df['AveragePrice'] < 100000, 'Houses price category'] = 'Low average price'
```

```python
df.loc[(df['AveragePrice'] >= 100000) & (df['AveragePrice'] < 300000), 'Houses price category'] = 'Middle average price'
```

```python
df.loc[df['AveragePrice'] >= 300000, 'Houses price category'] = 'High average price'
```

```python
df['Houses price category'].value_counts(dropna = True)
```

```python
# To create a categorical plot in seaborn using the house price categories created above

sns.set(style="ticks")
g = sns.catplot(x="SalesVolume", y="AveragePrice", hue="Houses price category", data = df)
```
